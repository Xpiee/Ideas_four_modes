```python
import numpy as np
import pandas as pd
import os
from main_utils_1 import mk_dirs
from shutil import copyfile
```

```python
## Read data for Prithila's EEG feature folder and copy it to the All Modalities folder

def process_eeg():
    foldr = ['MatbII', 'Virage']
    eegDrop = ['segment_length', 'label', 'Complexity', 'scaled_label', 'time_start', 'time_end', 'segment_number']

    for fldr in foldr:
        read_path = f'X:/All Modes/{fldr}/EEG/Unsync_EEG_features_Norm'
        eeg_path = f'X:/All Modes/All/{fldr}/Separate/EEG'
        mk_dirs(eeg_path)

        file_name = os.listdir(read_path)

        for sub in file_name:
            subPath = os.path.join(read_path, sub)
            savePath = os.path.join(eeg_path, sub)

            subDr = os.listdir(subPath)
            # extract exp number from subDr list
            exp_id = [x.replace('.csv', '') for x in subDr]
            exp_id = [x.replace('eeg_FE_norm_exp_', '') for x in exp_id]
            exp_id = set(exp_id)

            for exp in exp_id:
                # read eeg session file from the subject directory
                sessRead = os.path.join(subPath, f'eeg_FE_norm_exp_{exp}.csv')
                eegDF = pd.read_csv(sessRead)
                eegDF['Timestamp'] = eegDF['segment_number'].apply(lambda x: x - 10)
                eegDF.drop(columns=eegDrop, inplace=True)
                mk_dirs(savePath)
                copyFile = os.path.join(savePath, f'eeg_exp_{exp}.csv')
                eegDF.to_csv(copyFile, index=False)
```

```python
## Read data for Prithila's EEG feature folder and copy it to the All Modalities folder

def process_eeg_baseline():
    foldr = ['MatbII', 'Virage']
    eegDrop = ['segment_length', 'label', 'scaled_label', 'time_start', 'time_end', 'segment_number']

    for fldr in foldr:
        read_path = f'X:/All Modes/{fldr}/EEG/baseline'
        eeg_path = f'X:/All Modes/All/{fldr}/Baseline/EEG'
        mk_dirs(eeg_path)

        file_name = os.listdir(read_path)

        for sub in file_name:
            subPath = os.path.join(read_path, sub)
            savePath = os.path.join(eeg_path, sub)

            subDr = os.listdir(subPath)
            # extract exp number from subDr list
            exp_id = [x for x in subDr if '_baseline_' in x]
            exp_id = [x.replace('.csv', '') for x in exp_id]
            exp_id = [x.replace('eeg_FE_norm_baseline_', '') for x in exp_id]
            exp_id = set(exp_id)

            for exp in exp_id:
                # read eeg session file from the subject directory
                sessRead = os.path.join(subPath, f'eeg_FE_norm_baseline_{exp}.csv')
                eegDF = pd.read_csv(sessRead)
                eegDF['Timestamp'] = eegDF['segment_number'].apply(lambda x: x - 10)
                eegDF.drop(columns=eegDrop, inplace=True)
                mk_dirs(savePath)
                copyFile = os.path.join(savePath, f'eeg_baseline_exp_{exp}.csv')
                eegDF.to_csv(copyFile, index=False)

process_eeg_baseline()
```

```python
## Read data for Zunayed's Gaze feature folder and copy it to the All Modalities folder
def process_gaze():

    foldr = ['MatbII', 'Virage']
    gazeDrop = ['label', 'complexity', 'scaled_label', 'Time_start', 'Time_end', 'Segment_no']

    for fldr in foldr:
        read_path = f'X:/All Modes/{fldr}/eye-tracking/normalized/session_features_normalized'
        mod_path = f'X:/All Modes/All/{fldr}/Separate/Gaze'
        mk_dirs(mod_path)

        file_name = os.listdir(read_path)

        for sub in file_name:
            subPath = os.path.join(read_path, sub)
            savePath = os.path.join(mod_path, sub)

            subDr = os.listdir(subPath)

            # extract exp number from subDr list
            exp_id = [x.replace('.csv', '') for x in subDr]
            exp_id = [x.replace('exp', '') for x in exp_id]
            exp_id = [x.replace('level_', '') for x in exp_id]            
            exp_id = set(exp_id)

            for exp in exp_id:
                # read eeg session file from the subject directory
                if fldr == 'Virage':
                    sessRead = os.path.join(subPath, f'level_{exp}.csv')
                else:
                    sessRead = os.path.join(subPath, f'exp{exp}.csv')
                modDF = pd.read_csv(sessRead)
                modDF['Timestamp'] = modDF['Segment_no'].apply(lambda x: x - 10)
                modDF.drop(columns=gazeDrop, inplace=True)
                mk_dirs(savePath)
                copyFile = os.path.join(savePath, f'gaze_exp_{exp}.csv')
                modDF.to_csv(copyFile, index=False)

process_gaze()
```

```python
## Read data for Zunayed's Gaze feature folder and copy it to the All Modalities folder
def process_gaze_baseline():

    foldr = ['MatbII', 'Virage']
    gazeDrop = ['label', 'scaled_label', 'Time_start', 'Time_end', 'Segment_no']

    for fldr in foldr:
        read_path = f'X:/All Modes/{fldr}/eye-tracking/normalized/baseline'
        mod_path = f'X:/All Modes/All/{fldr}/Baseline/Gaze'
        mk_dirs(mod_path)

        file_name = os.listdir(read_path)

        for sub in file_name:
            subPath = os.path.join(read_path, sub)
            savePath = os.path.join(mod_path, sub)

            subDr = os.listdir(subPath)

            # extract exp number from subDr list
            exp_id = [x.replace('.csv', '') for x in subDr]
            exp_id = [x.replace('_baseline', '') for x in exp_id]
            exp_id = [x.replace('exp', '') for x in exp_id]
            exp_id = [x.replace('level_', '') for x in exp_id]            
            exp_id = set(exp_id)

            for exp in exp_id:
                # read eeg session file from the subject directory
                if fldr == 'Virage':
                    sessRead = os.path.join(subPath, f'level_{exp}_baseline.csv')
                else:
                    sessRead = os.path.join(subPath, f'exp{exp}_baseline.csv')
                modDF = pd.read_csv(sessRead)
                modDF['Timestamp'] = modDF['Segment_no'].apply(lambda x: x - 10)
                modDF.drop(columns=gazeDrop, inplace=True)
                mk_dirs(savePath)
                copyFile = os.path.join(savePath, f'gaze_baseline_exp_{exp}.csv')
                modDF.to_csv(copyFile, index=False)

process_gaze_baseline()
```

```python
def normalize_labels(values, actual_bounds, desired_bounds=(1, 9)):
    return [desired_bounds[0] + (x - actual_bounds[0]) * (desired_bounds[1] - desired_bounds[0]) / (actual_bounds[1] - actual_bounds[0]) for x in values]
```

```python
# REad and combine all the modalities together. All the best!

foldr = ['MatbII', 'Virage']
modalityList = ['ECG_EDA', 'EEG', 'Gaze']

desMin = 1
desMax = 9

for fldr in foldr:
    datasetReadPath = f'X:/All Modes/All/{fldr}/Separate'
    saveCsvPath = f'X:/All Modes/All/{fldr}/Combine'

    ecgedaPath = os.path.join(datasetReadPath, modalityList[0])
    eegPath = os.path.join(datasetReadPath, modalityList[1])
    gazePath = os.path.join(datasetReadPath, modalityList[2])

    subjectList = os.listdir(ecgedaPath)

    for sb in subjectList:
        try:

            exgsubPath = os.path.join(ecgedaPath, sb)
            eegsubPath = os.path.join(eegPath, sb)
            gazesubPath = os.path.join(gazePath, sb)

            expList = os.listdir(exgsubPath)
            expList = [x.replace('.csv', '') for x in expList]

            df_main = pd.DataFrame()
            for exp in expList:
                exgCsvRead = os.path.join(exgsubPath, f'{exp}.csv')
                eegCsvRead = os.path.join(eegsubPath, f'eeg_exp_{exp}.csv')
                gazeCsvRead = os.path.join(gazesubPath, f'gaze_exp_{exp}.csv')

                exgDF = pd.read_csv(exgCsvRead)
                eegDF = pd.read_csv(eegCsvRead)
                gzeDF = pd.read_csv(gazeCsvRead)

                # merging four modalities together for greater good.
                modDF = pd.merge(eegDF, exgDF, on ='Timestamp', how='right')
                modesDF = pd.merge(gzeDF, modDF, on='Timestamp', how='right')

                # Creating a combined session sheet for scaled labels
                df_main = df_main.append(modesDF)
                df_main = df_main.reset_index(drop=True)

            df_main.dropna(inplace=True)
            subMax = df_main['label'].max()
            subMin = df_main['label'].min()

            scLab = normalize_labels(df_main['label'].values, (subMin, subMax), (desMin, desMax))
            scLab = [round(x) for x in scLab]

            df_main['scaled label'] = scLab        
            mk_dirs(saveCsvPath)

            df_main.to_csv(os.path.join(saveCsvPath, '{}.csv'.format(sb)), index=False)

        except FileNotFoundError as e:
            print('Skipping this session as ', e)
            continue
```
