```python
# read baseline signals for each subject and plot them on a single line graph
baseline_path_matbii = r"X:/RealTimeSegment/MatbII/Raw/ECG_EDA_baseline/"
sess_baseline_matbii = r"X:/Four modes baseline/MatB-II_Clipped_Baseline"

listOfSubjects = os.listdir(baseline_path_matbii)

for subs in listOfSubjects:
    print(subs)
    
    subBasePath = os.path.join(baseline_path_matbii, subs)

    eda_baseline_file = os.path.join(subBasePath, "eda_baseline.csv")
    edaDF = pd.read_csv(eda_baseline_file, header=None, skiprows=1, names=['Timestamp', 'GSR Conductance CAL'])

    fig, ax = plt.subplots()
    fig.set_size_inches(18, 5)
    sn.set_style("whitegrid")

    img = sn.lineplot(x = edaDF.index, y=1000./edaDF['GSR Conductance CAL'], color="blue", label="3 Min. Baseline")

    if os.path.exists(os.path.join(sess_baseline_matbii, subs)):
        try:
            sessBaseDF = pd.DataFrame()
            sessBasePath = os.path.join(sess_baseline_matbii, subs)
            listOfSessions = os.listdir(sessBasePath)
            for ls_sess in listOfSessions:
                sessPath = os.path.join(sessBasePath, f'{ls_sess}')
                sessBaseDF = pd.read_csv(sessPath, usecols=['Timestamp', 'GSR Conductance CAL'])
                sessBaseDF.dropna(inplace=True)
                sessBaseDF.reset_index(inplace=True, drop=True)
                img = sn.lineplot(x = sessBaseDF.index, y=1000./sessBaseDF['GSR Conductance CAL'], label=ls_sess)

        except FileNotFoundError as e:
            # print(e)
            pass
        img.legend()
        plt.show()
    else: 
        img.legend()
        plt.show()

    plots = r'X:\Four modes baseline\Plots\MatbII'
    plt_sv = os.path.join(plots, '{}.png'.format(subs))
    img.figure.savefig(plt_sv, dpi=300)

    # break
```
